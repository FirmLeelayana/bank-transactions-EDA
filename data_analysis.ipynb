Let us now look for outliers in the dataset which may skew our analysis. We are doing this by the IQR score technique by drawing up a boxplot for the datapoints, and then removing those which lie beyond our boxplot.

```python
import seaborn as sns # for visualisation
sns.boxplot(x=data1['No Of Withdrawals']) #boxplot for number of withdrawals
```

```python
sns.boxplot(x=data1['Total amount Withdrawn']) #boxplot for total amount withdrawn
```

```python
#calculating the interquartile range for our numerical values for visualisation
q1 = data1.quantile(0.25)
q3 = data1.quantile(0.75)
IQR = q3 - q1
print(IQR)
```

```python
#Removing all our outliers for total amount withdrawn and total number of withdrawals from our dataset
data1 = data1[~((data1 < (q1 - 1.5 * IQR)) |(data1 > (q3 + 1.5 * IQR))).any(axis=1)]
data1.shape 
```

We can see that around 1000 of our datapoints were outliers, and so it was a good idea to remove these so that it does not skew our analysis and conclusions.

Task C - Manipulating the Dataset

Create a class that accepts the dataset file path as a parameter in the initialisation of the class, with certain implementation functions.

```python
from scipy import stats

class createDataset:
    '''
    Class that creates a panda Dataframe from a filepath argument linked to an Excel file, that can calculate the stdDev, 
    mean, and recognise outliers in the data.
    '''
    
    def __init__(self, filepath):
        '''
        Initialisation of the class.
        Arg: filepath, a string containing name of excel file in format .xlsx to be read from
        '''
        if not filepath.endswith('.xlsx') or type(filepath) != str: #must be in correct format
            raise TypeError('Please ensure that the excel filetype is in the .xlsx format, entering variable as a string.')
        self.filepath = filepath #set filepath
        self.dataset = pd.read_excel(filepath) #read file
        
    def display(self):
        '''
        Displays the first 5 excel data rows via panda's DataFrame.
        '''
        return self.dataset.head(5) 
    
    def stdDev(self, columnName):
        '''
        Calculates the standard deviation for a column.
        Arg: columnName, str, name of the column
        Returns: stdDev, float to 2 d.p, value of the standard deviation of that column
        '''
        if type(columnName) != str: #must be in correct format
            raise TypeError('Please ensure you enter a string.') 
        standardDev = self.dataset.std()[columnName]
        return ('%.2f' % standardDev) # two decimal places
    
    def mean(self, columnName):
        '''
        Calculates the mean for a column.
        Arg: columnName, str, name of the column
        Returns: mean, float to 2 d.p, value of the mean of that column
        '''
        if type(columnName) != str: #must be in correct format
            raise TypeError('Please ensure you enter a string.') 
        mean_val = self.dataset.mean()[columnName]
        return ('%.2f' % mean_val) # two decimal places
    
    def outliers(self):
        '''
        Identifies outliers for a dataset using a critial z-score of 3
        Arg: -
        Returns: list of rows that are outliers in the dataset
        '''
        row_panda = self.dataset[(np.abs(stats.zscore(self.dataset)) > 3).all(axis=1)] #those with z-scores less than 3
        row_list = row_panda.values.tolist()
        return row_list
```

```python
dataset_class = createDataset('ATMtransaction1.xlsx') #creating the class
```
